```python
import yfinance as yf
import datetime as dt
import os
```

```python
MACRO_TICKERS = ['^GSPC',
    'GC=F',
    'CL=F'
]
```

```python
startDate = dt.datetime(2023, 1, 1)
endDate = dt.datetime(2025, 1, 1)
```

```python
close_price_df = yf.download(MACRO_TICKERS,start=startDate,end=endDate)['Close']
close_price_df
```
